```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
```

```python
iris = pd.read_csv('iris.csv', index_col = 'Id')
```

```python
iris.columns
iris.head(5)
iris.describe()
```

```python
iris_species = iris.Species.unique()
iris_color = ['red', 'green', 'blue']
species_color = dict(zip(iris_species,iris_color))
species_color
```

```python
sepal_plot = plt.figure()
sepal_ax = sepal_plot.add_axes([1,1,1,1])

sepal_ax.set_xlabel('Sepal length(cm)')
sepal_ax.set_ylabel('Sepal width(cm)')
sepal_ax.set_title('Sepal length vs sepal width (cm)')

for key, value in species_color.items():
    sepal_ax.scatter(x=iris.SepalLengthCm[iris.Species==key],
         y=iris.SepalWidthCm[iris.Species==key], c=value, label=key)

sepal_ax.legend(loc=(1,0.7))

plt.show()
plt.close()
```

```python
petal_plot = plt.figure()
petal_ax = petal_plot.add_axes([1,1,1,1])

petal_ax.set_xlabel('Petal length(cm)')
petal_ax.set_ylabel('Petal width(cm)')
petal_ax.set_title('Petal length vs petal width (cm)')

for key, value in species_color.items():
    petal_ax.scatter(x=iris.PetalLengthCm[iris.Species==key],
         y=iris.PetalWidthCm[iris.Species==key], c=value, label=key)

petal_ax.legend(loc=(1,0.7))

plt.show()
plt.close()
```

```python
meow = sns.pairplot(iris, hue="Species")
meow.add_legend()

plt.show()
plt.close()
```
